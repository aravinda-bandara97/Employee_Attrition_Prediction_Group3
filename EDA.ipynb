### 9. Check Association between categorical features and the Response(Attrition) Using Chi-Square test

```python
from scipy.stats import chi2_contingency
# Assuming 'df' is your DataFrame
categorical_variables = df3.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_variables.remove("Attrition")
binary_response = "Attrition"

# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=["Categorical Feature", "Chi-Squared Statistic", "P-value"])

# Loop through each categorical feature
for feature in categorical_variables:
    # Create a contingency table between the categorical feature and the binary response
    contingency_table = pd.crosstab(df3[feature], df3[binary_response])

    # Perform the chi-squared test
    chi2, p, _, _ = chi2_contingency(contingency_table)

    # Append the results to the DataFrame
    results = results.append({"Categorical Feature": feature, "Chi-Squared Statistic": chi2, "P-value": p}, ignore_index=True)

# Set a significance level
alpha = 0.05

# Color significant associations at the 0.05 significance level
results["Significant"] = results["P-value"] < alpha

# Display the results
print(results)
```

* Based on Chi squared test most of categoricalmvariables are significantly associated with Employee attrition
